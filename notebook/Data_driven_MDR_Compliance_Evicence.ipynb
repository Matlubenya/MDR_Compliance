```python
# ðŸ”¹ Imports
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pdfplumber

# ðŸ”¹ Paths
DATA_DIR = "../data"
STATIM_FOLDERS = [os.path.join(DATA_DIR, "StatimA"), os.path.join(DATA_DIR, "StatimB")]
RITTER_FOLDERS = [os.path.join(DATA_DIR, "Ritter1"), os.path.join(DATA_DIR, "Ritter2")]

PROCESSED_DIR = "../processed"
RESULTS_DIR = "../results"

os.makedirs(PROCESSED_DIR, exist_ok=True)
os.makedirs(RESULTS_DIR, exist_ok=True)

print("Folders ready. Ready to process data.")
```

```python
# ðŸ”¹ Function to parse Statim TXT files
def parse_statim_file(filepath):
    with open(filepath, "r") as f:
        lines = f.readlines()
    record = {"Filename": os.path.basename(filepath)}
    for line in lines:
        if "Unit #" in line:
            record["Unit"] = line.split(":")[-1].strip()
        elif "Cycle Start" in line:
            record["Cycle Start"] = line.split()[-1].strip()
        elif "Sterilizing" in line and "C" in line:
            try:
                record["Sterilizing Temp (C)"] = float(line.split()[0])
            except:
                record["Sterilizing Temp (C)"] = None
    return record

# ðŸ”¹ Process Statim folders recursively
statim_records = []
for folder in STATIM_FOLDERS:
    for root, _, files in os.walk(folder):
        for file in files:
            if file.lower().endswith(".txt"):
                statim_records.append(parse_statim_file(os.path.join(root, file)))

# ðŸ”¹ Save CSV
statim_df = pd.DataFrame(statim_records)
statim_csv = os.path.join(PROCESSED_DIR, "statim_data_combined.csv")
statim_df.to_csv(statim_csv, index=False)
print(f"Statim data saved: {statim_csv}")

# ðŸ”¹ Plot: cycles per Statim unit
plt.figure(figsize=(8,5))
sns.countplot(data=statim_df, x="Unit")
plt.title("Number of Cycles per Statim Unit")
plt.savefig(os.path.join(RESULTS_DIR, "statim_cycles_per_unit.png"))
plt.show()
```
